load region dataset

```python
region_data=pd.read_csv("/content/noc_regions.csv")
```

```python
region_data.head()
```

```python
#merge dataset of region with olympic
olympic_data_new=pd.concat([olympic_data,region_data])
```

```python
#print the new data
olympic_data_new.head(10)
```

```python
#check shape
olympic_data_new.shape
```

```python
#print information about
olympic_data_new.info()
```

```python
#check columns
olympic_data_new.columns
```

```python
#check index
olympic_data_new.index
```

```python
#check for null values
olympic_data_new.isnull().any()
```

```python
olympic_data_new["Age"].fillna(olympic_data_new["Age"].mean(),inplace=True)
olympic_data_new["Height"].fillna(olympic_data_new["Height"].mean(),inplace=True)
olympic_data_new["Weight"].fillna(olympic_data_new["Weight"].mean(),inplace=True)

```

```python
olympic_data_new["Medal"].fillna(olympic_data_new["Medal"].mode()[0],inplace=True)
olympic_data_new["Sex"].fillna(olympic_data_new["Sex"].mode()[0],inplace=True)
olympic_data_new["City"].fillna(olympic_data_new["City"].mode()[0],inplace=True)
olympic_data_new["Sport"].fillna(olympic_data_new["Sport"].mode()[0],inplace=True)
olympic_data_new["Season"].fillna(olympic_data_new["Season"].mode()[0],inplace=True)
olympic_data_new["Name"].fillna(olympic_data_new["Name"].mode()[0],inplace=True)
olympic_data_new["Team"].fillna(olympic_data_new["Team"].mode()[0],inplace=True)
olympic_data_new["region"].fillna(olympic_data_new["region"].mode()[0],inplace=True)
olympic_data_new["Event"].fillna(olympic_data_new["Event"].mode()[0],inplace=True)
olympic_data_new["notes"].fillna(olympic_data_new["notes"].mode()[0],inplace=True)
olympic_data_new["ID"].fillna(olympic_data_new["ID"].mode()[0],inplace=True)
olympic_data_new["Games"].fillna(olympic_data_new["Games"].mode()[0],inplace=True)
olympic_data_new["Year"].fillna(olympic_data_new["Year"].mode()[0],inplace=True)
```
